```python
import pandas as pd
import os
import gpxpy
```

```python
files = os.listdir(path = './gpx/')
files
```

```python
# loop through files, merge into one dataframe
def parsegpx(f):
    points2 = []
    with open(f, 'r') as gpxfile:
        # print f
        gpx = gpxpy.parse(gpxfile)
        for track in gpx.tracks:
            for segment in track.segments:
                for point in segment.points:
                    dict = {'Timestamp' : point.time,
                            'Latitude' : point.latitude,
                            'Longitude' : point.longitude,
                            'Elevation' : point.elevation,
                            'Time': point.time
                            }
                    points2.append(dict)
    return points2 
```

```python
df2 = pd.concat([pd.DataFrame(parsegpx("gpx/" + f)) for f in os.listdir("gpx")], keys=files)
df2.sample(250)
```
