```python
html_text
```

```python
#pull name, date, activity stats for each entry on page
#skip if all values aren't present

activity_details = []

for html in html_text:
    bs = BeautifulSoup(html, 'html.parser')
    container = bs.find_all('div', attrs={"class":"entry-container"})
    for a in container:
        try:
            name = a.find_all('a', attrs={'class':"entry-athlete"})[0].get_text().strip('\n').split('\n')[0]
            date = a.find_all('time', attrs={'class':"timestamp"})[0].get_text().strip('\n')
            dist = a.find_all('li', attrs={'title':"Distance"})[0].get_text()
            pace = a.find_all('li', attrs={'title':"Pace"})[0].get_text()
            dur = a.find_all('li', attrs={'title':"Time"})[0].get_text()
            activity_details.append([name, date, dist, pace, dur])
        except:
            pass
activity_details
```

```python
#convert to dataframe and name columns
adam_df = pd.DataFrame(activity_details, columns=['name', 'date', 'distance', 'pace', 'duration'])
adam_df
```

```python
#convert date to datetime object
adam_df['date'] = adam_df['date'].str.replace('at', '')

adam_df.date=pd.to_datetime(adam_df.date)
adam_df
```

```python
adam_df['duration_sum'] =adam_df['duration'].str.split(' ')
adam_df
```

```python
#create column for duration in minutes
def total_duration(x):
    min_sum = 0
    for item in x:
        min_sum += convert_to_minutes(item)
    return min_sum

def convert_to_minutes(time):
    time_type = time[-1];
    if time_type == "h":
        return int(time[0:-1]) * 60
    if time_type =="m":
        return int(time[0:-1])
    else:
        return 0


adam_df['duration_sum']=adam_df['duration_sum'].map(total_duration)
adam_df
```

```python
#remove /mi from pace and convert to fraction
adam_df['pace'] =adam_df['pace'].str.replace('/mi', '')
def express_pace_decimal(row):
    try:
        minutes_part = float(row.split(":")[0])
        seconds_part = float(row.split(":")[1])
        minute_fraction = seconds_part / 60
        return (minutes_part +  minute_fraction)
    except:
        return '0'
adam_df['pace_fraction'] = adam_df['pace'].map(express_pace_decimal) 
adam_df
```

```python
#filter the data for run dates
adam_summary = adam_df.sort_values(by=['date'])
adam_summary = adam_summary[(adam_df['date'] > '2016-02-11 09:32:00') & (adam_summary['date'] <= '2016-04-14 01:40:00')]

adam_summary
```

```python
adam_summary['distance'] = adam_summary['distance'].str.replace('mi', '')
adam_summary
```

```python
#split dattime into 2 columns so I can grouby day
adam_summary['day'] = [d.date() for d in adam_summary['date']]
adam_summary['time'] = [d.time() for d in adam_summary['date']]

#convert to distance to number
adam_summary['distance']= pd.to_numeric(adam_summary['distance'], errors='coerce')
new = adam_summary.groupby(['day', 'name']).agg(['sum', 'mean'])[['distance', 'pace_fraction', 'duration_sum']].reset_index()
new.columns = new.columns.map('_'.join)
new.drop(['distance_mean', 'pace_fraction_sum', 'duration_sum_mean'], axis =1, inplace=True)
# new = new['day_', 'name_', 'distance_sum', 'pace_fraction_mean', 'duration_sum_sum']
new.columns = ['day', 'name', 'total_distance', 'avg_pace', 'duration']
adam_run_summary = new
adam_run_summary
```
