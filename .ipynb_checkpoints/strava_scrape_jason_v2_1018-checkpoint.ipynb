```python
date_stat = list(zip([i for i in dates if len(i.split()) == 6], [i for i in activities if len(i.split()) >= 6]))
date_stat
```

```python
df = pd.DataFrame(date_stat)
df.columns=['date', 'stats']
df
```

```python
df['date'] = df['date'].str.replace('at', '')

df.date=pd.to_datetime(df.date)
df
```

```python
#splits stats column into individual values
def split_stats(x):
    vals= x.split()
    dist = vals[0]
    pace = vals[2]
    duration = vals[4:6]
    return pd.Series([dist, pace, duration], index = ('distance', 'pace', 'duration'))
    
jr_stats = df['stats'].apply(split_stats)

jr_stats['date']= df['date']
# jr_stats['duration'] = jr_stats['duration'].str.replace('h', '')
jr_stats
```

```python
jr_stats['duration']
```

```python
def express_pace_decimal(row):
    try:
        minutes_part = float(row.split(":")[0])
        seconds_part = float(row.split(":")[1])
        minute_fraction = seconds_part / 60
        return (minutes_part +  minute_fraction)
    except:
        return '0'
jr_stats['pace'] = jr_stats['pace'].map(express_pace_decimal) 
```
