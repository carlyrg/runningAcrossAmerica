```python
#pull name, date, activity stats for each entry on page
#skip if all values aren't present

activity_details = []

for html in html_text:
    bs = BeautifulSoup(html, 'html.parser')
    container = bs.find_all('div', attrs={"class":"entry-container"})
    for a in container:
        try:
            name = a.find_all('a', attrs={'class':"entry-athlete"})[0].get_text().strip('\n').split('\n')[0]
            date = a.find_all('time', attrs={'class':"timestamp"})[0].get_text().strip('\n')
            dist = a.find_all('li', attrs={'title':"Distance"})[0].get_text()
            pace = a.find_all('li', attrs={'title':"Pace"})[0].get_text()
            dur = a.find_all('li', attrs={'title':"Time"})[0].get_text()
            activity_details.append([name, date, dist, pace, dur])
        except:
            pass
activity_details
```

```python
#convert to dataframe and name columns
jason_df = pd.DataFrame(activity_details, columns=['name', 'date', 'distance', 'pace', 'duration'])
jason_df
```

```python
#convert date to datetime object
jason_df['date'] = jason_df['date'].str.replace('at', '')
jason_df.date=pd.to_datetime(jason_df.date)
jason_df
```

```python
jason_df['duration_sum'] =jason_df['duration'].str.split(' ')
jason_df
```

```python
#create column for duration in minutes
def total_duration(x):
    min_sum = 0
    for item in x:
        min_sum += convert_to_minutes(item)
    return min_sum

def convert_to_minutes(time):
    time_type = time[-1];
    if time_type == "h":
        return int(time[0:-1]) * 60
    if time_type =="m":
        return int(time[0:-1])
    else:
        return 0

jason_df['duration_sum'] = jason_df['duration_sum'].map(total_duration)
jason_df
```

```python
#remove /mi from pace and convert to decimal 
jason_df['pace'] =jason_df['pace'].str.replace('/mi', '')

def express_pace_decimal(row):
    try:
        minutes_part = float(row.split(":")[0])
        seconds_part = float(row.split(":")[1])
        minute_fraction = seconds_part / 60
        return (minutes_part +  minute_fraction)
    except:
        return '0'
jason_df['pace_fraction'] = jason_df['pace'].map(express_pace_decimal) 
jason_df['pace_fraction'] = pd.to_numeric(jason_df['pace_fraction'], errors='coerce')

# jason_df
jason_df.dtypes
```

```python
#filter the data for run dates
jason_summary = jason_df.sort_values(by=['date'])
jason_summary = jason_summary[(jason_summary['date'] > '2016-03-15 16:55:00') & (jason_summary['date'] <= '2016-05-23 03:07:00')]

jason_summary
```

```python
jason_summary['distance'] = jason_summary['distance'].str.replace('mi', '')
jason_summary
```

```python
#split dattime into 2 columns so I can grouby day
jason_summary['day'] = [d.date() for d in jason_summary['date']]
jason_summary['time'] = [d.time() for d in jason_summary['date']]

#convert to distance to number
jason_summary['distance']= pd.to_numeric(jason_summary['distance'], errors='coerce')
new = jason_summary.groupby(['day', 'name']).agg(['sum', 'mean'])[['distance', 'pace_fraction', 'duration_sum']].reset_index()
new.columns = new.columns.map('_'.join)
new.drop(['distance_mean', 'pace_fraction_sum', 'duration_sum_mean'], axis =1, inplace=True)
# new = new['day_', 'name_', 'distance_sum', 'pace_fraction_mean', 'duration_sum_sum']
new.columns = ['day', 'name', 'total_distance', 'avg_pace', 'duration']
jason_run_summary = new
jason_run_summary
```
