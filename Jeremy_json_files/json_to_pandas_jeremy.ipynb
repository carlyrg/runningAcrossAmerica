```python
import json
import pandas as pd
import os
import glob
from tqdm import tqdm
import re
from pandas.io.json import json_normalize
from datetime import datetime
import numpy as np
```

```python
json_ll = glob.glob("/Users/snowr/ds_python1/running-America/vru_retry/Jeremy_json_files/jb_latlng/*.json")
json_as = glob.glob("/Users/snowr/ds_python1/running-America/vru_retry/Jeremy_json_files/jb_activity_stats/*.json")

assert(len(json_ll) == len(json_as))

for_df = []

keys = [re.findall(r'\d{5,}' , x.split("/")[-1])[0] for x in json_ll]

# goal: join the files based on their id

def get_my_dict_from_the_file(the_file_path):
    # returns a dictionary
    with open(the_file_path, 'r') as f:
        data = eval(f.read())
        return data

# below will give a list of dicts within for_df (list). I'll need to concat them all using pd.concat(axis=1?)

for key in tqdm(keys):
    # get the matching ll file
    # get the matching activity file
    # get the date later!
    matching_ll_file = [x for x in json_ll if key in x][0]
    matching_as_file = [x for x in json_as if key in x][0]
    #print(key, matching_ll_file, matching_as_file)
    ll_dict = get_my_dict_from_the_file(matching_ll_file)
    activity_dict = get_my_dict_from_the_file(matching_as_file)
    merged_dict = {**ll_dict, **activity_dict}
    merged_dict['activity_id'] = key
    print(merged_dict.keys())
    for_df.append(merged_dict)
    
    
pd.DataFrame.from_records(merged_dict).shape
```

```python
out = pd.concat([pd.DataFrame.from_records(x) for x in for_df], axis=0)
out
```
