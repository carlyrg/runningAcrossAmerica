```python
#extract activity id, stats url, geo data url and store in lists
#activity stats json

def get_json_links():
    
    jr_activity = []
    jr_activity_id = activities
    jr_latlng = []


    for a in activities:
        #return json page for activity
        jr_activity.append("{}/streams?stream_types%5B%5D=altitude&stream_types%5B%5D=heartrate&stream_types%5B%5D=cadence&stream_types%5B%5D=temp&stream_types%5B%5D=distance&stream_types%5B%5D=grade_smooth&stream_types%5B%5D=time&stream_types%5B%5D=grade_adjusted_distance&_=1538018467153".format(a))

        #return json page for latlng
        jr_latlng.append("{}/streams?stream_types%5B%5D=timer_time&stream_types%5B%5D=latlng&_=1537842990619".format(a))

    #create and save files
    with open('jr_activities_list.txt', 'w') as f:
        for item in jr_activity:
            f.write("%s\n" % item)
    with open('jr_latlng_list.txt', 'w') as f:
        for item in jr_latlng:
            f.write("%s\n" % item)
    with open('jr_activity_id.txt', 'w') as f:
        for item in jr_activity_id:
            f.write("%s\n" % item)
            
    return jr_activity, jr_activity_id, jr_latlng

jr_activity, jr_activity_id, jr_latlng = get_json_links() # This gets all the links and writes a text file

get_json_links()
```

```python
#save JSON files locally for activity stats
def get_json_activity_data(rows_for_iteration):
    try:
        #save the json_locally
        
        browser = webdriver.Chrome() 
        browser.get("https://www.strava.com/login") #navigate to the page
        browser.implicitly_wait(5)

                    # login = browser.find_element_by_xpath('//*[@id="global-nav"]/div/ul/li[5]/a').click()

        #sign-in
        username = browser.find_element_by_id('email')
        username.clear()
        username.send_keys(username_c)

        password = browser.find_element_by_id('password')
        password.clear()
        password.send_keys(strava_pw)

        browser.find_element_by_id("login-button").click()

        #wait to load
        time.sleep(5) 
        
        
        for row in tqdm(rows_for_iteration):
            try:
                save_json_filename = re.findall('activities\/\d{5,}', row)[0].split("/")[-1] + "_s_jr.json"
#                 save_json_filename = re.findall('activities\/\d{5,}', row)[0].split("/")[-1] + "_ll_jr.json"

                browser.get(row)
                pre = browser.find_element_by_tag_name("pre").text
                data = json.loads(pre)
                with open(save_json_filename, 'w') as f:
                    f.write(json.dumps(data))
                time.sleep(2)
            except Exception as e:
                print(row, str(e)[:200])
                time.sleep(2)
    
    finally:
        browser.quit()
        
get_json_activity_data(jr_activity)
# get_json_activity_data(jr_latlng)
```

```python
#get date of activity
def get_date(activities):
    try:
        browser = webdriver.Chrome() 

        jr_date = dict()
        for x in activities:
            try:
                browser.get(x)
                browser.implicitly_wait(5)
                date = browser.find_element_by_class_name("timestamp").text
                activity = x.split("/")[-1]
                jr_date.update({activity:date})
            
            except Exception as e:
                jr_date.update({"na":"na"})
                print(x, str(e)[:300])
                time.sleep(2)    
    finally:
        browser.quit()
        
    return jr_date

jr_date = get_date(jr_activity_id)
```

```python
#convert to JSON file and save
jr_date_json = json.dumps(jr_date)
jr_date_json

with open('jr_date.json', 'w') as f:
    json.dump(jr_date, f)
```
